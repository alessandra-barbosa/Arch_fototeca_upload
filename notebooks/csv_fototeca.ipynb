### Informazione Foto da upload

```python
# Informazione Foto da upload
nome_cartella =r'C:\Users\Utente77\Desktop\Foto Cilone _da organizzare\ASS DONNE CGIL 0606'
evento='Assemblea nazionale donne Spi-Cgil' #Titolo Assegnato
anno= '2012' #Periodo
Fotografo='Mina Celoni'
extensione_file=['.JPG','.jpg']
Descrizione=evento+"_"+anno #Titolo Originale
Id =""
Altra_denominazione=""
Antroponimi=""
Descrittori=""
Localizzazione=""
Luogo=""
Dimensioni_unità =""
Inquadratura=""
Stato_Conservazione=""
Collocazione=""
Archivio=""
Classificazione=""
```

#### Crea una nuova cartella

```python

nuova_cartella=nome_cartella+'/'+evento
# try:
os.mkdir(nuova_cartella)
# except FileExistsError as e:
    # # input('cancellare cartella y or n?')
    # # if 'y':
    #     shutil.rmtree(nuova_cartella)
    #     os.mkdir(nuova_cartella)
    # # else:
    #print('ok')
```

### Cancellare folder

```python
 shutil.rmtree(nuova_cartella)
```

### Rinomeando le foto

```python
# Crea una lista di foto
lista_arquivos = os.listdir(nome_cartella)



# Renomea file e copia alla nuova cartella
for arquivo in lista_arquivos:
    if arquivo.lower().endswith(".jpg"): # solo file jpg
        arquivo_novo=anno+'_'+evento+'_'+arquivo
        shutil.copy2(nome_cartella+"/"+arquivo,nuova_cartella+'/'+arquivo_novo)
    
```

### Creando File .CSV

```python

lista_final= os.listdir(nuova_cartella)
df = pd.DataFrame(data=lista_final, columns =['nome_file'])
df['Id']=Id
df['Titolo Originale']=Descrizione
df['Titolo Assegnato']=evento
df['Altra Denominazione']=Altra_denominazione
df['Antroponimi']=Antroponimi
df['Descrittori']=Descrittori
df['Localizzazione']=Localizzazione
df['Periodo']=anno
df['Luogo']=Luogo
df['Dimensioni/unità ']=Dimensioni_unità
df['Fotografo']=Fotografo
df['Inquadratura']=Inquadratura
df['Stato Conservazione']=Stato_Conservazione
df['Collocazione']=Collocazione
df['Archivio']=Archivio
df['Classificazione']=Classificazione



df.to_csv(nuova_cartella+'/'+'file.csv', index=False)  
```
